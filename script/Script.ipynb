```python
mark=[]
for index,x in enumerate(df.columns):
    split = x.split("_")
    if split[-1]=='timestamp' or split[-1]=='complete':
        mark.append(index)
addition = df.iloc[:, mark[0]+1:mark[1]].sum(axis=1)
addition
```

```python
df.columns[mark[1]].split("_complete")[0]
```

```python
def addNewColumn(score,name):
    df.insert(loc=mark[1]+1,column=df.columns[mark[1]].split("_complet")[0]+name,value=score,allow_duplicates=False)   
```

```python
addNewColumn(addition,'-total-score')
```

```python
df[df.columns[11:22]]
```

```python
selected = df.columns[mark[0]+1:mark[1]]
selected
```

```python
def subscore(x):
    selectedCols = []
    for i in x:
        selectedCols.append(selected[i-1])
    return df[selectedCols].sum(axis=1)
subscore([4,5,6,7,8])
```

```python
addNewColumn(subscore([1, 2, 5, 7, 8, 9, 11, 12, 13]),'-subscore-WM')
```
